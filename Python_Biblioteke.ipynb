```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(1, 50, 100) #linspace(start, end, num_points)
y = np.sin(x)

#linijski graf
plt.plot(x, y)
```

```python
#grafik rasipanja
plt.scatter(x, y)
```

```python
plt.plot(x, y)
plt.scatter(x, y)
```

```python
plt.plot(x, y)
fig = plt.figure()
plt.scatter(x,y)
```

```python
plt.plot(x, y, 'x-', color = 'red', alpha = .5, label = 'the data')

plt.title('The Title')
plt.xlabel('x label')
plt.ylabel('y label')

plt.legend(loc = 'best')
plt.grid()
plt.savefig('../test_fig.png')
```

```python
t = np.linspace(1, 100, 100)

x1 = np.sin(20./(2*np.pi)*t)
x2 = np.sin(40./(2*np.pi)*t)
x3 = x1 + x2

plt.figure()
plt.subplot(2, 1, 1)
plt.plot(t, x1, label = 'x1')
plt.plot(t, x2, label = 'x2')
plt.plot(t, x3, label = 'x3')
plt.legend(loc = 'best')

plt.subplot(2, 1, 2)
x1hist = plt.hist(x1, density = 'true', bins = 25, alpha =.25, label = 'x1')
plt.legend(loc = 'best')
```

```python
x1 = np.random.randn(1000)
x2 = 2.0*np.random.randn(1000)*2

x1h = plt.hist(x1, bins = 50, alpha = .25)
x2h = plt.hist(x2, bins = 50, alpha = .25)
```
